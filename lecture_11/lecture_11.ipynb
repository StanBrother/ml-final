## Unsupervised Learning

Unsupervised სწავლება არის ML-ის ტიპი, სადაც მოდელი სწავლობს "ადამიანის ზედამხედველობის" გარეშე. Supervised სწავლებისგან განსხვავებით, unsupervised სწავლების ML მოდელებს მონაცემებს ეტიკეტების გარეშე ვაწვდით და ვაძლევთ უფლებას, რომ იპოვონ შაბლონები და კანონზომიერებები დამატებითი ინსტრუქციების ან ხელმძღვანელობის გარეშე.

როგორც სახელი მიანიშნებს, unsupervised სწავლებისას ვიყენებთ self-learning ალგორითმებს - ისინი სწავლობენ ეტიკეტების/ლეიბლების გარეშე. მოდელს ვაწვდით მონაცემებს ეტიკეტების გარეშე და მას უწევს თავად დააგენერიროს წესები და ინფორმაციის სტრუქტურირება მოახდინოს მონაცემების მსგავსებებზე, განსხვავებებზე და შაბლონებზე დაყრდნობით.

Unsupervised სწავლების ალგორითმები გამოიყენება კომპლექსური დამუშავების ამოცანებისთვის. მაგალითად, უზარმაზარი მონაცემების კლასტერებად დაყოფა. ისინი შეგვიძლია გამოვიყენოთ მონაცემების მანამდე ამოუცნობი შაბლონების იდენტიფიცირებისთვის და დავიხმაროთ მონაცემების კატეგორიზაციისთვის სასარგებლო მახასიათებლების იდენტიფიცირებისას.

ვთქვათ, გვაქვს ბევრი მონაცემი ამინდის შესახებ. Unsupervised სწავლების ალგორითმი გადაივლის ამ მონაცემებზე და ამოიცნობს შაბლონებს მონაცემთა წერტილებისთვის. მაგალითად, მას შეუძლია მონაცემების დაჯგუფება ტემპერატურის ან სხვა მსგავსი შაბლონის მიხედვით. მიუხედავად იმისა, რომ თავად ალგორითმს არ "ესმის" შაბლონები ჩვენ მიერ მოწოდებული წინა ინფორმაციის საფუძველზე, ჩვენ შეგვიძლია დავაკვირდეთ ამ დაჯგუფებულ მონაცემებს და ვცადოთ მათი კლასიფიკაცია. მაგალითად, შეიძლება, რომ სხვადასხვა ტემპერატურული ჯგუფი წარმოადგენს სხვადასხვა სეზონს ან რომ ნაპოვნი ამინდის შაბლონები შეიძლება დაიყოს სხვადასხვა ტიპის ამინდად, როგორიცაა წვიმა, სეტყვა ან თოვლი.

### კლასტერიზაცია

კლასტერიზაცია არის დაუმუშავებელი, არაეტიკეტირებული მონაცემების შესწავლისა და მსგავსებების ან განსხვავებების საფუძველზე ჯგუფებად (ან კლასტერებად) დაყოფის ტექნიკა. იგი გამოიყენება სხვადასხვა აპლიკაციაში, მათ შორის მომხმარებელთა სეგმენტაციაში, თაღლითობის გამოვლენასა და გამოსახულების ანალიზში. კლასტერიზაციის ალგორითმები მონაცემებს ბუნებრივ ჯგუფებად ყოფს დაუხარისხებელ მონაცემებში მსგავსი სტრუქტურების ან შაბლონების პოვნის გზით.

არსებობს რამდენიმე კლასტერიზაციის ალგორითმი:

* **Exclusive** - მონაცემები დაჯგუფებულია ისე, რომ ერთი მონაცემთა წერტილი შეიძლება არსებობდეს მხოლოდ ერთ კლასტერში. ხანდახან მოიხსენიება, როგორც "hard" კლასტერიზაცია. Exclusive კლასტერიზაციის მაგალითია K-means ალგორითმი, რომელიც ანაწილებს მონაცემთა წერტილებს მომხმარებლის მიერ განსაზღვრულ K რაოდენობის კლასტერში.
* **Overlapping** - მონაცემები დაჯგუფებულია ისე, რომ ერთი მონაცემთა წერტილი შეიძლება არსებობდეს ორ ან მეტ კლასტერში წევრობის სხვადასხვა ხარისხით. ამას ასევე უწოდებენ "soft" კლასტერიზაციას.
* **Hierarchical** - მონაცემები დაყოფილია განსხვავებულ კლასტერებად მსგავსებაზე დაყრდნობით, რომლებიც შემდეგ კიდევ ერწყმიან ერთმანეთს. ისინი ორგანიზებულია მათი იერარქიული ურთიერთობების საფუძველზე. არსებობს იერარქიული კლასტერიზაციის ორი ძირითადი ტიპი: აგლომერაციული (agglomerative) და გამყოფი (divisive). ამ მეთოდს ასევე მოიხსენიებენ როგორც HAC-იერარქიული კლასტერული ანალიზი.
* **Probabilistic** - მონაცემები დაჯგუფებულია კლასტერებად ალბათობის მიხედვით, რომ თითოეული მონაცემთა წერტილი მიეკუთვნება კონკრეტულ კლასტერს. ეს მიდგომა განსხვავდება სხვა მეთოდებისგან, რომლებიც აჯგუფებენ მონაცემთა წერტილებს მათი მსგავსების საფუძველზე.

#### K-Means

K-means არის პოპულარული კლასტერიზაციის ალგორითმი, რომელიც გამოიყენება მონაცემთა წერტილების ნაკრების K ჯგუფად (კლასტერად) დასაყოფად. თავდაპირველად ის ირჩევს K რაოდენობის შემთხვევით წერტილს კლასტერების ცენტრებად. მონაცემთა წერტილებს ანიჭებს უახლოეს ცენტროიდს. შემდგომ კლასტერში შემავალი წერტილებისთვის ითვლის საშუალოს, რომ ახალი ცენტროიდი მონახოს. ეს პროცესი მეორდება მანამ, სანამ ცენტროიდები მნიშვნელოვნად აღარ შეიცვლება, რაც ნიშნავს, რომ კლასტერები სტაბილურია.

სამწუხაროდ, არ არსებობს ოპტიმალური K, რომელიც ყველა მონაცემთა ნაკრებისთვის გამოგვადგება, თუმცა არსებობს "იდაყვის" (elbow) მეთოდი, რომელიც გვეხმარება K-ს ამორჩევაში. ამისთვის საჭიროა რამდენიმე K-ს მნიშვნელობისთვის მოვახდინოთ კლასტერიზაცია და გამოვთვალოთ მონაცემთა წერტილებისთვის კვადრატული მანძილების ჯამი (Within-Cluster Sum of Square - WCSS) უახლოესი კლასტერის ცენტრამდე და ეს გამოვსახოთ გრაფიკზე. იდეა იმაში მდგომარეობს, რომ წერტილი, სადაც WCSS-ის შემცირების სიჩქარე მკვეთრად იცვლება (იდაყვის მსგავსი ფორმა აქვს გრაფიკს) კარგი არჩევანია კლასტერების რაოდენობისთვის.

````python
wcss = []
for n_cluster in range(1, 11):
    kmeans = KMeans(
        n_clusters=n_cluster, init="k-means++", max_iter=300, n_init=10, random_state=1
    )
    kmeans.fit(customers_df[["Annual Income (k$)", "Spending Score (1-100)"]])
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker="o")
plt.title("Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
````

გრაფიკის მიხედვით შეგვიძლია 5 კლასტერი გვქონდეს:

````python
kmeans = KMeans(n_clusters=5, init="k-means++", max_iter=300, n_init=10, random_state=0)
clusters = kmeans.fit_predict(
    customers_df[["Annual Income (k$)", "Spending Score (1-100)"]]
)

plt.figure(figsize=(10, 6))
sns.scatterplot(
    customers_df,
    x="Annual Income (k$)",
    y="Spending Score (1-100)",
    hue=clusters,
    palette="viridis",
    s=100,
)
plt.legend(title="Cluster")
plt.show()
````

### ასოციაციის წესები

ასოციაციის წესების mining არის წესებზე დაფუძნებული მიდგომა, რათა გამოავლინოს საინტერესო ურთიერთობები მონაცემთა წერტილებს შორის მონაცემთა დიდ ნაკრებებში. unsupervised სწავლების ალგორითმები ეძებენ ხშირ if-then ასოციაციებს, რომლებსაც ასევე უწოდებენ წესებს, რათა აღმოაჩინონ კორელაციები მონაცემებში და მონაცემთა ობიექტებს შორის განსხვავებული კავშირები.

ეს მეთოდი ყველაზე ხშირად გამოიყენება სავაჭრო კალათების ან ტრანზაქციის მონაცემთა ნაკრების გასაანალიზებლად, რათა წარმოაჩინოს რამდენად ხშირად ხდება გარკვეული საქონლის ერთობლივი შეძენა. ეს ალგორითმები ეძებენ კლიენტების შესყიდვის შაბლონებს და აქამდე "დამალულ" ურთიერთობებს პროდუქტებს შორის, რაც ხელს უწყობს სარეკომენდაციო სისტემების დახვეწას. ხშირად არის ხოლმე ზოგიერთ ონლაინ მაღაზიის საიტზე შემდეგი სახის სექციები: "Frequently bought together" და "People who bought this item also bought".

### განზომილების შემცირება
